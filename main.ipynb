```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact
from IPython.display import display

# 1. 載入資料（請確認路徑正確）
df = pd.read_csv('整理後.csv')

# 2. 擷取所有文字類別欄位，並排除「時間戳記」
all_text_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
cols = [c for c in all_text_cols if c != '時間戳記']

# 3. 繪圖函式：顯示交叉表、長條圖、以及針對每個類別畫圓餅圖
def plot_cross(x_col, y_col):
    cross_tab = pd.crosstab(df[x_col], df[y_col])
    display(cross_tab)  # 顯示交叉表

    # 長條圖
    plt.figure(figsize=(10, 6))
    cross_tab.plot(kind='bar', stacked=False)
    plt.title(f'{x_col} vs {y_col}', fontsize=16)
    plt.xlabel(x_col)
    plt.ylabel('人數')
    plt.xticks(rotation=0)
    plt.legend(title=y_col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # 分群圓餅圖
    for val in cross_tab.index:
        plt.figure(figsize=(6, 6))
        cross_tab.loc[val].plot(kind='pie', autopct='%1.1f%%', startangle=90)
        plt.title(f'{x_col} = {val} 的 {y_col} 分布', fontsize=14)
        plt.ylabel('')
        plt.tight_layout()
        plt.show()

# 4. 建立互動介面
interact(
    plot_cross,
    x_col=widgets.Dropdown(options=cols, description='X 軸欄位：'),
    y_col=widgets.Dropdown(options=cols, description='Y 軸欄位：')
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact
from IPython.display import display

# 1. 載入資料
df = pd.read_csv('整理後.csv')

# 2. 準備下拉選單的欄位清單（排除「時間戳記」）
all_text_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
cols = [c for c in all_text_cols if c != '時間戳記']

# 3. 繪圖函式：針對每個 Z 的類別做交叉表和長條圖
def plot_three(x_col, y_col, z_col):
    # 3.1 顯示三維交叉表（MultiIndex 版）
    ct3 = pd.crosstab(
        index=[df[z_col], df[x_col]],
        columns=df[y_col]
    )
    display(ct3)

    # 3.2 分組繪圖：對每個 Z 的值，畫一張 X vs Y 的長條圖
    for z_val, sub in df.groupby(z_col):
        ct2 = pd.crosstab(sub[x_col], sub[y_col])
        plt.figure(figsize=(8, 5))
        ct2.plot(kind='bar', stacked=False)
        plt.title(f'{z_col} = {z_val} → {x_col} vs {y_col}', fontsize=14)
        plt.xlabel(x_col)
        plt.ylabel('人數')
        plt.xticks(rotation=0)
        plt.legend(title=y_col, bbox_to_anchor=(1.02, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

# 4. 建立三個下拉選單的互動介面
interact(
    plot_three,
    x_col = widgets.Dropdown(options=cols, description='X 軸：'),
    y_col = widgets.Dropdown(options=cols, description='Y 軸：'),
    z_col = widgets.Dropdown(options=cols, description='分組(Z)：')
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact
from IPython.display import display

# 1. 載入資料
df = pd.read_csv('整理後.csv')

# 2. 準備下拉選單的欄位清單（排除「時間戳記」）
all_text_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
cols = [c for c in all_text_cols if c != '時間戳記']

# 3. 繪圖函式：對每個 Z 的值，先畫 X vs Y 的長條圖，再對每個 X 值畫 Y 的圓餅圖
def plot_three_with_pie(x_col, y_col, z_col):
    # MultiIndex 三維交叉表
    ct3 = pd.crosstab(
        index=[df[z_col], df[x_col]],
        columns=df[y_col]
    )
    display(ct3)

    # 逐一對 Z 群組作圖
    for z_val, sub in df.groupby(z_col):
        ct2 = pd.crosstab(sub[x_col], sub[y_col])

        # 3.1 長條圖
        plt.figure(figsize=(8, 5))
        ct2.plot(kind='bar', stacked=False)
        plt.title(f'{z_col} = {z_val} → {x_col} vs {y_col}', fontsize=14)
        plt.xlabel(x_col)
        plt.ylabel('人數')
        plt.xticks(rotation=0)
        plt.legend(title=y_col, bbox_to_anchor=(1.02, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

        # 3.2 圓餅圖：對該 Z 群組裡每個 X 值各自畫
        for x_val in ct2.index:
            plt.figure(figsize=(5, 5))
            ct2.loc[x_val].plot(
                kind='pie',
                autopct='%1.1f%%',
                startangle=90
            )
            plt.title(f'{z_col}={z_val}，{x_col}={x_val} 的 {y_col} 分布', fontsize=12)
            plt.ylabel('')
            plt.tight_layout()
            plt.show()

# 4. 建立三個下拉選單的互動介面
interact(
    plot_three_with_pie,
    x_col = widgets.Dropdown(options=cols, description='X 軸：'),
    y_col = widgets.Dropdown(options=cols, description='Y 軸：'),
    z_col = widgets.Dropdown(options=cols, description='分組(Z)：')
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from ipywidgets import widgets, interact
from IPython.display import display, clear_output

# 1. 讀入資料
df = pd.read_csv('整理後.csv')

# 2. 準備下拉選單欄位（排除「時間戳記」）
text_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
choices = [c for c in text_cols if c != '時間戳記']

# 3. 定義互動式繪圖與檢定函式
def auto_chi2(x_col, y_col, z_col):
    clear_output()  # 每次選擇都先清除上一次輸出
    print(f"── 整體 {x_col} vs {y_col} 卡方檢定 ──")
    overall = pd.crosstab(df[x_col], df[y_col])
    chi2, p, dof, exp = chi2_contingency(overall)
    print(f"χ² = {chi2:.2f},  p = {p:.4f} → {'顯著(p<0.05)' if p<0.05 else '不顯著'}")
    display(overall)
    
    # 針對每個 Z 群組
    for z_val, sub in df.groupby(z_col):
        print(f"\n── {z_col} = {z_val} 的卡方檢定 ──")
        ct = pd.crosstab(sub[x_col], sub[y_col])
        chi2, p, _, _ = chi2_contingency(ct)
        print(f"χ² = {chi2:.2f},  p = {p:.4f} → {'*顯著*' if p<0.05 else '不顯著'}")
        display(ct)
        
        # 長條圖
        plt.figure(figsize=(6,4))
        ct.plot(kind='bar', stacked=False)
        plt.title(f'{z_col}={z_val}: {x_col} vs {y_col}')
        plt.xticks(rotation=0)
        plt.tight_layout()
        plt.show()
        
        # 圓餅圖
        for x_val in ct.index:
            plt.figure(figsize=(4,4))
            ct.loc[x_val].plot(
                kind='pie', autopct='%1.1f%%', startangle=90
            )
            plt.title(f'{z_col}={z_val}, {x_col}={x_val}')
            plt.ylabel('')
            plt.tight_layout()
            plt.show()

# 4. 建立三個下拉選單
interact(
    auto_chi2,
    x_col = widgets.Dropdown(options=choices, description='X 軸：'),
    y_col = widgets.Dropdown(options=choices, description='Y 軸：'),
    z_col = widgets.Dropdown(options=choices, description='分組(Z)：')
)
```
